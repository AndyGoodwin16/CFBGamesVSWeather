```python
#Import dependencies.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
#Read in csv file.
data = pd.read_csv("final.csv")
```

```python
#Create a new column with temperature in F instead of C.
data["tavgF"] = data["tavg"]*9/5 + 32
#Drop None values from column.
data1 = data.dropna(subset=["tavgF"])
#Create scatter plot showing relationship between Total Score and Temperature.
x_values = data1["tavgF"]
y_values = data1["Total Score"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,10))
plt.scatter(x_values, y_values, color = 'b')
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Average Temperature (F)")
plt.ylabel("Total Score")
plt.title("Total Score vs Temperature")
plt.annotate(line_eq,(0,10),fontsize=15,color="r");
tcorr = data["tavgF"].corr(data["Total Score"], method = 'pearson')
print(f"The correlation (Pearson) between Total Score and Temperature is {round(tcorr, 2)}.")
```

```python
#Create a new column with wind speed in mph instead of km/h.
data["wspdmph"] = data["wspd"]/1.60934
#Drop None values from column.
data2 = data.dropna(subset=["wspdmph"])
#Create scatter plot showing relationship between Total Score and Wind Speed.
x_values = data2["wspdmph"]
y_values = data2["Total Score"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,10))
plt.scatter(x_values, y_values, color = 'r')
plt.plot(x_values,regress_values,"b-")
plt.xlabel("Wind Speed (mph)")
plt.ylabel("Total Score")
plt.title("Total Score vs Wind Speed")
plt.annotate(line_eq,(21,10),fontsize=15,color="b");
wcorr = data["wspdmph"].corr(data["Total Score"], method = 'pearson')
print(f"The correlation (Pearson) between Total Score and Wind Speed is {round(wcorr, 2)}.")
```

```python
#Create a new column with pressure normalized to 0 as sea level.
data["pres0"] = data["pres"] - 1013.25
#Drop None values from column.
data3 = data.dropna(subset=["pres0"])
#Create scatter plot showing relationship between Total Score and Pressure.
x_values = data3["pres0"]
y_values = data3["Total Score"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,10))
plt.scatter(x_values, y_values, color = 'g')
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Pressure (millibar, normalized to 0)")
plt.ylabel("Total Score")
plt.title("Total Score vs Pressure")
plt.annotate(line_eq,(-20,7),fontsize=15,color="r");
pcorr = data["pres0"].corr(data["Total Score"], method = 'pearson')
print(f"The correlation (Pearson) between Total Score and Pressure is {round(pcorr, 2)}.")
```

```python
#Create list of Precipitation data to iterate over.
prcp_list = data["prcp"].tolist()
#Loop through all Precipitation data and set all values <10 to None.
for i in range(len(prcp_list)):
    if prcp_list[i] < 10:
        prcp_list[i] = None
#Add new Precipitation data list to dataframe.
data["prcpM10"] = prcp_list
#Create scatter plot showing relationship between Total Score and Precipitation.
data4 = data.dropna(subset=["prcpM10"])
x_values = data4["prcpM10"]
y_values = data4["Total Score"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,10))
plt.scatter(x_values, y_values, color = 'orange')
plt.plot(x_values,regress_values,"b-")
plt.xlabel("Precipitation (mm, more than 10 mm)")
plt.ylabel("Total Score")
plt.title("Total Score vs Precipitation")
plt.annotate(line_eq,(100,17),fontsize=15,color="b");
prcorr = data["prcpM10"].corr(data["Total Score"], method = 'pearson')
print(f"The correlation (Pearson) between Total Score and Precipitation is {round(prcorr, 2)}.")
```

```python
#Create list of Snaw Fall data to iterate over.
snow_list = data["snow"].tolist()
#Loop through all Snow Fall data and set all values <25 to None.
for i in range(len(snow_list)):
    if snow_list[i] < 25:
        snow_list[i] = None
#Add new Precipitation data list to dataframe.
data["snowM25"] = snow_list
#Create scatter plot showing relationship between Total Score and Snow Fall.
data5 = data.dropna(subset=["snowM25"])
x_values = data5["snowM25"]
y_values = data5["Total Score"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(10,10))
plt.scatter(x_values, y_values, color = 'purple')
plt.plot(x_values,regress_values,"r-")
plt.xlabel("Snow Fall (mm, more than 25 mm)")
plt.ylabel("Total Score")
plt.title("Total Score vs Snow Fall")
plt.annotate(line_eq,(100,25),fontsize=15,color="r");
scorr = data["snowM25"].corr(data["Total Score"], method = 'pearson')
print(f"The correlation (Pearson) between Total Score and Snow Fall is {round(scorr, 2)}.")
```
